```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk import word_tokenize
from datetime import datetime
    
df = pd.read_csv('trumptwit-dropped.csv', encoding='latin-1')
df.created_at = df.created_at.apply(lambda x: datetime.strptime(x,'%m/%d/%y %H:%M'))
```

```python
def wordfreq(df, top_N = 100):
    a = re.sub('[^A-Za-z0-9 ]+', '', df['text'].str.lower().str.cat(sep=' '))
    stop = stopwords.words('english')
    stop.append('realdonaldtrump')
    text = ' '.join([word for word in a.split() if word not in stop])
    words = nltk.tokenize.word_tokenize(text)
    word_dist = nltk.FreqDist(words)
    rslt = pd.DataFrame(word_dist.most_common(top_N),
                        columns=['Word', 'Frequency'])
    return rslt
```

```python
dday = datetime(2016, 6, 16)
eday = datetime(2016, 11, 8)

rslt_pre = wordfreq(df[df.created_at < dday])
rslt_election = wordfreq(df[(df.created_at >= dday) & (df.created_at < eday)])
rslt_president = wordfreq(df[df.created_at >= eday])
rslt_pre['flag'] = 'precandidacy'
rslt_election['flag'] = 'election'
rslt_president['flag'] = 'president'
rslt_pre.to_csv('rslt_pre.csv')
rslt_election.to_csv('rslt_election.csv')
rslt_president.to_csv('rslt_president.csv')
```
